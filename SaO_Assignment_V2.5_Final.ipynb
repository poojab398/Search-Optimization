# I. Problem Definition

**Background:**   
    A real-world problem of delivery planning for LPG Ditribution Company (**SaO Gas Ltd**) has been given, that operates its 25 tanker lorries from 4 depots. The goal is to minimise the overall cost of delivery for the distributor under given constraints. The Optimization problems will be enumerated in the sections below.

## 2.1 Euclidean distance

In mathematics, the Euclidean distance between two points in Euclidean space is the length of a line segment between the two points. It can be calculated from the Cartesian coordinates of the points using the Pythagorean theorem, therefore occasionally being called the Pythagorean distance. 

The distance between two objects that are not points is usually defined to be the smallest distance among pairs of points from the two objects.

### 2.1.1 Distance formulas

#### One dimension

The distance between any two points on the real line is the absolute value of the numerical difference of their coordinates. Thus if $p$ and $q$ are two points on the real line, then the distance between them is given by:
$d(p,q)$ = $|p - q|$

A more complicated formula, giving the same value, but generalizing more readily to higher dimensions, is:
$$d(p,q) = \sqrt{(p-q)^2}$$ 

In this formula, squaring and then taking the square root leaves any positive number unchanged, but replaces any negative number by its absolute value.


#### Two dimensions

In the Euclidean plane, let point p have Cartesian coordinates $( p 1 , p 2 )$ and let point q have coordinates $( q 1 , q 2 )$. Then the distance between $p$ and $q$ is given by:

$$d(p,q) = \sqrt{(p_1 - q_1)^2 + (p_2 - q_2)^2}$$ 

This can be seen by applying the Pythagorean theorem to a right triangle with horizontal and vertical sides, having the line segment from $p$ to $q$ as its hypotenuse. The two squared formulas inside the square root give the areas of squares on the horizontal and vertical sides, and the outer square root converts the area of the square on the hypotenuse into the length of the hypotenuse.

It is also possible to compute the distance for points given by polar coordinates. If the polar coordinates of $p$ are $( r , \theta )$ and the polar coordinates of $q$ are $( s , \psi )$ , then their distance is:

$$d(p,q) = \sqrt{r^2 + s^2 - 2rs\cos(\theta - \psi)}$$

When $p$ and $q$ are expressed as complex numbers in the complex plane, the same formula for one-dimensional points expressed as real numbers can be used:

$$d(p,q) = |p - q|$$


### 2.1.2 Shortest path between nodes

**Dijkstra's algorithm** (or Dijkstra's Shortest Path First algorithm, SPF algorithm) is an algorithm for finding the shortest paths between nodes in a graph, which may represent, for example, road networks. The Dijkstra algorithm uses labels that are positive integers or real numbers, which are totally ordered. It can be generalized to use any labels that are partially ordered, provided the subsequent labels (a subsequent label is produced when traversing an edge) are monotonically non-decreasing. This generalization is called the generic Dijkstra shortest-path algorithm.

### Shortest path visualisation
#### Definition
The shortest path problem can be defined for graphs whether undirected, directed, or mixed. It is defined here for undirected graphs; for directed graphs the definition of path requires that consecutive vertices be connected by an appropriate directed edge.

Two vertices are adjacent when they are both incident to a common edge. A path in an undirected graph is a sequence of vertices
$$P = (v_1,v_2,...,v_n) \in V \times V \times ... \times V $$

such that $v_i+1$ is adjacent to $1 \leqslant i < n$

The problem is also sometimes called the single-pair shortest path problem, to distinguish it from the following variations:
- The single-source shortest path problem, in which we have to find shortest paths from a source vertex v to all other vertices.
- The single-destination shortest path problem, in which we have to find shortest paths from all vertices in the directed graph to a single destination vertex v. This can be reduced to the single-source shortest path problem by reversing the arcs in the directed graph.
- The all-pairs shortest path problem, in which we have to find shortest paths between every pair of vertices v, v'.

These generalizations have significantly more efficient algorithms than the simplistic approach of running a single-pair shortest path algorithm on all relevant pairs of vertices. 


### 2.1.3 Distance matrix

A distance matrix is a square matrix (two-dimensional array) containing the distances, taken pairwise, between the elements of a set. Depending upon the application involved, the distance being used to define this matrix may or may not be a metric.

If there are $N$ elements, this matrix will have size $NxN$. In graph-theoretic applications the elements are more often referred to as points, nodes or vertices.

## 2.2.1 Greedy search solution

**Greedy search** is a variation of the A* algorithm, where `g(node)` is set to zero, so that only `h(node)` is used to evaluate suitable paths. In this way, the algorithm always selects the path that has the lowest heuristic value or estimated distance (or cost) to the goal.

The Greedy  approach is called **Best-first Search**. Best-first search is an instance of the general **TREE-SEARCH or GRAPH-SEARCH** algorithm in which a node is selected for expansion based on an evaluation function, $f(n)$. 

The evaluation function is construed as a cost estimate, so the node with the lowest evaluation is expanded first. The implementation of best-first graph search is identical to that for uniform-cost search, except for the use of $f$ instead of $g$ to order the priority queue.

Most best-first algorithms include as a component of $f$ a heuristic function, denoted $h(n)$:  <br>

<div style="text-align: center">$h(n)$ = estimated cost of the cheapest path from the state at node $n$ to a goal state.</div>

***Commonly Best-first search is based upon Cost, but we are basing it upon Distance so at each step we move to the nearest node which needs fuel.***

## 2.2.2 Genetic Algorithm
**[Sharma N, 2020]**
Genetic Algorithms are inspired by nature to "improve" populations. Among organisms, each parent passes down **genes** to their children, which affects the ability of the children to survive. It is an analogy of this principle of biological genetic evolution that is the basis for the heuristic optimization method called "Genetic Algorithm" (GA). In this algorithm, the "genes" are decision variables.

#### Definitions
* Gene: basic genetic element
* Chromosome: collection of genes
* Allele: The values a gene can take (eg. 0 or 1)

  For example, if $s$ is a binary string of length 4, examples of chromosome are `1010`, `0110` etc. Genes are the bits in the string, and alleles are 0 or 1.
* Fitness: The objective function, $F(s)$ that we want to *maximize*
* Generation: A single iteration of the GA algorithm

### a. Characteristics of Genetic Algorithms
Binary genetic algorithms are ones where the decision variable is a binary string. The primary operations in genetic algorithms are: 
* **Crossover**: This involves switching locations of substrings within a binary string between two strings.
* **Mutation**: This involves changing the value of one gene in the binary string on a probability distribution.

### b. Outline of Genetic Algorithm
1. Randomly generate an initial “ population ” $\{s^{curr}_j\}  \forall j= 1,…,M$.
2. Compute fitness $F(s^{curr}_j) \forall j= 1,…,M$
3. Generate "children" which are the members of the population in the next generation by applying:
  * Selection of pairs of parents (by roulette or tournament) randomly influenced by $F(s^{curr}_j)$
  * Crossover
  * Mutation
4. The population of children becomes a population of parents for the next iteration and goto step 2 unless a stopping criterion has been satisfied.

#### Crossover (using Roulette)
In order to generate $N_c$ children, we perform the steps:
* Randomly select one parent with a probability that is proportional to its fitness $F(s_i)$.
$$p_i = [\frac{F(s_i)}{\sum_{k=1}^{N_p}F(s_k)}]$$
* Then randomly select a second parent also with a probability that is proportional to the fitness of the parent.
* For one point crossover, pick a crossover point. This can be done randomly although there may be some restrictions
    * Multipoint cross over randomly selects multiple crossover points, effectively cropping multiple genes between paretns.
* Then switch the protions of the binary strings that occur before the crossover point between the parents.

#### Crossover (using Tournament)
* Let all parents have an equal probability of being selected (e.g. $p_i = 1/M$, where $M$ is number of individuals in the population).
* Pick two parents $s_i$ and $s_k$ at random.
* Select the parent with the higher fitness to be the first parent.
* Repeat the process starting with step 1 to select the second parent

#### Mutation
* Mutation is applied after crossover and produces new characteristics. WIthout mutation, there would be some values we would never reach.
* It produces incremental random changes in the offspring by randomly changing allele values. Typically, a mutation flips a value with $S$ is a binary string.


### 2.2.3 Ant Colony Optimization Algorithm

**[Mirjalili S. 2019]**   
**Swarm intelligence** is a unique way oof solving real world problem by considering the social behaviour of insects and other animals. Specifically, ants have inspired new methods and techniques to tackle the problem and the most successful and well-known approach was the ant colony optimization. 

**Ant colony optimization (ACO)** takes inspiration from social interaction of ants and thus using foraging behaviour. These ants deposit a chemical called pheromone on the ground enabling other ants to follow the most favourable path to reach the destination.

**Ant Colony Optimization on a network:**  
The process starts by simulating number of artificial ants on the network allowing to encode the problem itself. Each customer location in the network is considered as vertex and road line between customers is called as edge. And these edges and vertices can be read by ants to find the shortest route possible. 

ACO is an iterative process and at each iteration, ants start building the solution by moving from vertex to vertex on the network and limiting themselves not to visit the vertex that has already been visited. This phenomenon of ants is based on stochastic mechanism that is biased by pheromone deposit and evaporation rate. If location $i$ has not already been visited, it can be selected by considering the probability associated with pheromone deposit with the edge $(i,j)$.

The first colony optimization algorithm is known as Ant system. Its main characteristic is that, at each iteration the value of pheromone is updated by all the $m$ ants.  The pheromone value ${\tau_{ij}}$ is updated by the following equation. 

$${\tau_{ij} < -(1 - \rho) . \tau_{ij} + \sum_{k=1}^{m} \varDelta \tau_{ij}^{k}}$$

Where ${\rho}$ = evaporation rate
&emsp;$m$ =Number of ants
&emsp; ${\varDelta \tau_{ij}^{k}}$ = Quantity of pheromone laid


$${\varDelta \tau_{ij}^{k}} = \begin{cases}
Q/L_k, \quad \text{if ant $k$ used edge(i, j) in tour},\\
0, \quad \text{otherwise}
\end{cases}$$

Where $Q$ = Constant  
&emsp; &emsp; ${L_k}$ = length of tour travelled by ant $k$

Once the ant selects the next city to be travelled by stochastic mechanism, when ant is in the city $i$ and has calculated the partial solution ${S^P}$, and decides to go the city $j$ using following probability given below.


$$p_{ij}^{k} = \begin{cases}
\frac{\tau_{ij}^{\alpha} \eta_{ij}^{\beta}}{\sum_{cil \in N({S^P})} \tau_{ij}^{\alpha} \eta_{ij}^{\beta}} \quad \text{if $c_{ij} \in N({S^P})$}\\
0, \quad \text{otherwise}
\end{cases}$$

Where $N({S^P})$ = Set of feasible components   
&emsp; &emsp; $\alpha$ and $\beta$ = Controls the relative importance of the pheromone versus heuristic information $\eta_{ij}$   

&emsp; &emsp; $\eta_{ij}  = \frac{1}{$d_{ij}}$ &emsp; &emsp; $d_{ij}$ = distance between cities $i$ and $j$

## 3.5 Network & Shortest Path Visualization

We will now create a plot showing the road map of *Optilandia* using the `networkx` library, so we will create a function to do this.

```python
# These need to be accessed outside the scope of the function so they can be used elsewhere

# creating a graph using networkx
G = nx.Graph()

# iterrows - iterate over dataframe rows and provide row number in k and row data in v
# which will extract the (x, y) coords of each location and place them in dictionary pos
pos = {k:v.values for k, v in df_loc[['x','y']].iterrows()}

def show_network():
    # construct network edge objects to be added to the graph
    edges  = [(id1,id2,dist_eucl[id1,id2]) for _,(id1,id2) in df_link.iterrows()]
    
    # add_nodes_from() takes an iterable container of node tuples and adds them to graph
    G.add_nodes_from(range(len(df_loc)))
    
    # add_weighted_edges_from() takes a container of edge values and adds them to graph
    G.add_weighted_edges_from(edges)

    # Extract node ids where df_loc field is_depot is true
    depot_loc_ids = np.where(df_loc.is_depot)[0]
    # Extract node ids where df_loc field is_customer is true
    cust_loc_ids = np.where(df_loc.is_customer)[0]
    
    # Plotting the graph
    plt.figure(figsize=(20,10))
    # draw the network graph using (x, y) coords in pos
    nx.draw(G,with_labels=False,pos=pos,node_size=50)
    # redraw the nodes in the depot_loc_ids list with large red dots
    nx.draw_networkx_nodes(G,pos,nodelist=depot_loc_ids,node_color='r',node_size=600,alpha=1);
    # draw labels on the depot nodes showing their id numbers
    nx.draw_networkx_labels(G,pos,{k:k for k in depot_loc_ids});
    # redraw the nodes in the cust_loc_ids list with small green dots
    nx.draw_networkx_nodes(G,pos,nodelist=cust_loc_ids,node_color='g',node_size=100,alpha=1);
```

## 3.7 Devising Validation Sets (Problem 1)

We then create a function named `validate_solution(sol, lorries_dict)`, which loops from all the solutions and performs validation checks. The function returns `False` if the solution is valid or a short message describing the failed check(s) for the below parameters:
- One journey per lorry.
- Each lorry must start from its home depot.
- Lorries cant be loaded beyond its capacity.
- Lorries can not deposit gas more than it has onboard.

```python
def validate_solution(sol, lorries_dict):
    
    # all lorry ids from the solutions
    all_ids = [s['lorry_id'] for s in sol]
    
    # non-existent lorry_id
    for id in all_ids:
        if id not in lorries_dict:
            return f"Non-existent lorry id {id}."
        
    # one journey per lorry
    if len(all_ids) != len(set(all_ids)): return "Multiple journeys per lorry."
    
    # for each journey in the solution
    for s in sol:
        
        # empty journey is ok, skip further checks
        if len(s['loc']) == 0: continue
        
        # each lorry must start from its home depot
        if int(s['lorry_id'].split('-')[0]) != int(s['loc'][0][0]):
            return f"Lorry {s['lorry_id']} should start journey from its home depot."
        
        # lorry can't be loaded beyond its capacity at any point
        if np.any(np.cumsum([c[1] for c in s['loc']]) > lorries_dict[s['lorry_id']]['capacity']):
            return f"Lorry {s['lorry_id']} exceeds its capacity."
        
        # lorry can't deliver more gas than it has; le-12 to allow for roundoff errors
        if np.any(np.cumsum([c[1] for c in s['loc']]) < -1e-12):
            return f"Lorry {s['lorry_id']} dropped off more gas than it had onboard."
        
    return False
```

```python
def validate_solution2(sol, lorries_dict, depot_id_list, dist_graph):
    
    # all lorry ids from the solutions
    all_ids = [s['lorry_id'] for s in sol]
    
    # non-existent lorry_id
    for id in all_ids:
        if id not in lorries_dict:
            return f"Non-existent lorry id {id}."
        
    # one journey per lorry
    if len(all_ids) != len(set(all_ids)): return "Multiple journeys per lorry."
    
    # for each journey in the solution
    for s in sol:
        
        # empty journey is ok, skip further checks
        if len(s['loc']) == 0: continue
        
        # each lorry must start from its home depot
        if int(s['lorry_id'].split('-')[0]) != int(s['loc'][0][0]):
            return f"Lorry {s['lorry_id']} should start journey from its home depot."
        
        # each lorry must end at a depot
        if int(s['loc'][-1][0]) not in depot_id_list:
            return f"Lorry {s['lorry_id']} should end journey at a depot."
        
        # each lorry can make 5 stops
        if len(s['loc']) > 6:
            return f"Lorry {s['lorry_id']} shouldn't make more than 5 stops."
        
        # lorry can't be loaded beyond its capacity at any point
        if np.any(np.cumsum([c[1] for c in s['loc']]) > lorries_dict[s['lorry_id']]['capacity']):
            return f"Lorry {s['lorry_id']} exceeds its capacity."
        
        # lorry can't deliver more gas than it has
        if np.any(np.cumsum([c[1] for c in s['loc']]) < -1e-12):
            return f"Lorry {s['lorry_id']} dropped off more gas than it had onboard."
        
        # lorry can't travel more than 250 miles
        mileage = 0
        i = 1
        if len(s['loc']) > 1:
            while i < len(s['loc']):
               mileage += dist_graph[int(s['loc'][i-1][0]),int(s['loc'][i][0])]
               i += 1
            if mileage > 250:
                return f"Lorry {s['lorry_id']} travelled more than 250 miles."
        
    return False
```

#### Nearest Neighbour (NN)

Here we create a function to find the nearest customer that needs fuel (unfilled customer), starting from the home depot.

```python
# Create a function to find the nearest customer that still needs fuel
# Takes 1D array of distances from a particular node indexed by node position &
# a copy of the same arrow sorted by distance
# Returns node id of nearest customer that needs fuel
def nearest_customer(row_distance, sorted_distance):
    counter = 1
    customer_found = False
    
    # loop whilst customer hasn't been found and index is less than the number of nodes
    while counter < len(row_distance) and customer_found == False:
        index = 0
        is_nearest = False

        # loop to find the nearest node
        while is_nearest == False and index < len(sorted_distance):
            if sorted_distance[counter] == row_distance[index]:
                is_nearest = True
            else:
                index += 1

        # loop through cust_dict1 to check whether the node is a customer and needs fuel
        for i in cust_dict1:
            if index == i and cust_dict1[i]['fuel_needed'] > 0:
                customer_found = True
        if customer_found == False:
            counter += 1
    
    if counter == len(row_distance):
        return 0
    else:
        return index;
```

Next, for each lorry, we create a solution list (route path) for each journey of unfilled customer(s) that needs fuel (based on the Shortest Path calculation solution above). The lorry starts with full capacity from the depot, can visit the first unfilled customer, drop the fuel as needed for that customer, move to next unfilled customer (if the lorry has fuel has equal to greater than the cumulative demand as each customer node).

If the lorry has less fuel than the customer fuel needed for subsequent customer nodes on the journey, it will deliver all fuel and return to the nearest depot for refuel.

```python
# Set index for solution file
sol_index = 0

# Set end statement
all_customers_full = False

# While end statement not True
while all_customers_full == False:
    # for each lorry
    for lorry in lorries_dict1:
        if sol_index > 24:
            sol_index = 0

        # if lorry empty return to depot and fuel up
        if lorries_dict1[lorry]['level'] == 0:
            lorries_dict1[lorry]['position'] = lorries_dict1[lorry]['depot']
            lorries_dict1[lorry]['level'] = lorries_dict1[lorry]['capacity']
            solution_list[sol_index]['loc'].append([lorries_dict1[lorry]['position'], lorries_dict1[lorry]['level']])

        # find closest unfilled customer
        # Make a 1D array of distances from a particular node
        row_distance = dist_graph[lorries_dict1[lorry]['position']]
        # Make a copy of this array sorted by distance
        sorted_distance = np.sort(row_distance)
        # Find nearest unfilled customer
        unfilled_customer = nearest_customer(row_distance = row_distance, sorted_distance = sorted_distance)
        if unfilled_customer == 0:
            break;

        # Set lorry position to unfilled customer
        lorries_dict1[lorry]['position'] = unfilled_customer

        # If the lorry has less fuel than the customer fuel needed, deliver all and set lorry level to 0
        if lorries_dict1[lorry]['level'] < cust_dict1[unfilled_customer]['fuel_needed']:
            cust_dict1[unfilled_customer]['fuel_needed'] -= round(lorries_dict1[lorry]['level'], 2)
            cust_dict1[unfilled_customer]['level'] += round(lorries_dict1[lorry]['level'], 2)
            lorries_dict1[lorry]['fuel_delivered'] += round(lorries_dict1[lorry]['level'], 2)
            solution_list[sol_index]['loc'].append([unfilled_customer, round(lorries_dict1[lorry]['level'] * -1, 2)])
            lorries_dict1[lorry]['level'] = 0
        else: # If the lorry has equal or greater than fuel needed, deliver amount and set fuel needed to 0
            lorries_dict1[lorry]['level'] -= round(cust_dict1[unfilled_customer]['fuel_needed'], 2)
            cust_dict1[unfilled_customer]['level'] += round(cust_dict1[unfilled_customer]['fuel_needed'], 2)
            lorries_dict1[lorry]['fuel_delivered'] += round(cust_dict1[unfilled_customer]['fuel_needed'], 2)
            solution_list[sol_index]['loc'].append([unfilled_customer, round(cust_dict1[unfilled_customer]['fuel_needed'] * -1, 2)])
            cust_dict1[unfilled_customer]['fuel_needed'] = 0

        # Increment sol_index
        sol_index += 1
    
    # Check end statement
    sum_fuel_needed = 0
    
    for i in cust_dict1:
        sum_fuel_needed += cust_dict1[i]['fuel_needed']
        
    if sum_fuel_needed == 0:
        all_customers_full = True
```

```python
#Function to perform mutation
def main():
    
    # The use of system time as the seed will ensure a genuinely random initial population
    random.seed(datetime.datetime.now())

    # generating a population with size of 400
    pop = toolbox.population(n=400)

    #copb  is the crossover probability
    #copb= 0.8
    # MUTPB is the probability for mutating an individual
    MUTPB = 0.25
    
    print("Start of evolution")
    
    # Evaluate fitness of each individual in the population
    fitnesses = list(map(toolbox.evaluate, pop))
    
    
    for ind, fit in zip(pop, fitnesses):
        ind.fitness.values = fit
            
    print("  Evaluated %i individuals" % len(pop))

    # Extracting all the fitnesses of each individual 
    fits = [ind.fitness.values[0] for ind in pop]

    # Variable keeping track of the number of generations
    p = 0
    
    # Begin the evolution
    while p < 4000:
        # A new generation
        p = p + 1
        print("-- Generation %i --" % p, end=" ")
        
        # Select the next generation individuals
        offspring = toolbox.select(pop, len(pop))

        # Clone the selected individuals
        offspring = list(map(toolbox.clone, offspring))
        
        #Mutation process
        for mutant in offspring:

            # mutate an individual with probability MUTPB
            if random.random() < MUTPB:
                toolbox.mutate(mutant)
                del mutant.fitness.values
    
        # Evaluate the individuals with an invalid fitness
        invalid_ind = [ind for ind in offspring if not ind.fitness.valid]
        fitnesses = map(toolbox.evaluate, invalid_ind)
        for ind, fit in zip(invalid_ind, fitnesses):
            ind.fitness.values = fit
        
        # The population is entirely replaced by the offspring
        pop[:] = offspring
        
        # Gather all the fitnesses in one list and print the stats
        fits = [ind.fitness.values[0] for ind in pop]
        
        # print("----------------")
        print("  Min %s" % min(fits))
            
    print("-- End of (successful) evolution --")
    
    best_ind = tools.selBest(pop, 1)[0]
    print("Best individual is %s" % (best_ind))
    print("Best fitness value is %s" % (best_ind.fitness.values))
```

### 4.3 Ant Colony Optimization Technique for Optimization Problem 1

We chose Ant Colony optimization as one of the intelligent algorithms to implement Problem 1 of the assignment. 

Travelling salesman problem was based as an inspiration to develop the methodology for this problem, each depot location was considered as source point and a set of nearest customer locations would be the cities traversed along the journey.  

To create the optimal routes from each of the depot locations, ACO algorithm was implemented as explained below:
1.	Different number of ants travel from one location to destination using different paths. So, we can set the number of ants starting from a location as $N$.
2.	Ants deposit pheromone on the paths they navigate to reach the destination. We initialised a matrix of pheromone deposits, it would be of the same shape as the distance matrix with the co-ordinates corresponding to the customer locations. We initialized the pheromone matrix with small values for all the locations. 
3.	Next the paths are explored from each depot location to its respective customers. The decision to choose the next location to move in the journey is done using the Probability measure ${(P_{i,j}^{k}(t))}$ which is evaluated using the formula below:

$$p_{ij}^{k} = \begin{cases}
\frac{\tau_{ij}^{\alpha}(t) \eta_{ij}^{\beta}(t)}{\sum_{u \in N_i^{k}(t)} \tau_{iu}^{\alpha}(t) \eta_{iu}^{\beta}(t)} \quad \text{if $j \in N_i^k(t)$}\\
0, \quad \text{if $j \notin N_i^k(t)$}
\end{cases}$$

&emsp;&emsp;${\tau}$ - represents the pheromone deposit from location $i$ and $j$ in path $t$  
&emsp;&emsp;${\eta}$ - represents the cost factor of distance between $i$ and $j$ in path $t$  
&emsp;&emsp;${\alpha}$ and ${\beta}$ -  act as weights on pheromone and distance, respectively. 

4.	Ants keep moving to different locations along the journey using the above rule, if we chose 10 ants to lead the journeys, we will have 10 paths at the end of this group of ants travelling generation. 
5.	Initially we have same amount of pheromone deposited on all paths, so the choices are made based on the distances, however on their way back ants deposit pheromone on the paths they travelled, the pheromone amount is calculated as:

$$\frac{1}{distance_{ij}}$$

&emsp;&emsp;where ${distance_{ij}}$ = Distance between nodes $i$ and $j$

&emsp;&emsp;This formula helps add more weightage to the shorter routes between 2 locations and higher probability of choosing this path in next iteration. 

6.	Along with pheromone deposit, another property of ACO is the pheromone decay, which is done to identify the least traversed paths. We just multiply the pheromone matrix by the decay rate and the path visited the least number of times will have very small value.
7.	This process is continued for $n$ number of iterations and the final optimal route is determined by the shortest path resulting from the list of journeys traversed. 
8.  This process is carried out on 4 depot locations and delivery scheduling is done on the journeys generated using above  explained ACO algorithm.

Now we write the ACO algorithm to generate optimal paths for the lorries to navigate for gas delivery. 
The ACO function takes in 4 parameters -
    1. Distance Matrix
    2. Customer List for the respective depot
    3. Number of Iterations
    4. Number of Ants
    5. Alpha and Beta values

```python
#Ant Colony Optimization Implementation for Problem 1:

#Initialization of all the parameters:

#input distance matrix
def ant_colony_opt(distance_matrix,cust_list,iterations,ants,alpha,beta):
    distance = distance_matrix

    #number of ants for iteration
    m=ants

    #number of iterations
    n=iterations

    #no of locations
    nodes=len(cust_list)

    #Pheromone matrix
    pheromone = np.ones(distance.shape)/len(distance)

    # rate of decay
    decay = 0.5

    # cost of path i and j
    eta =1/distance
    eta[eta==inf] =0

    #pheromone factor
    alpha =alpha

    #cost factor
    beta = beta

    #intializing probability features array to zero
    prob_feature = np.zeros(nodes)     

    #intializing cummulative probability array to zeros
    cum_prob = np.zeros(nodes)            


    #intializing the path of the ants with size route(ants,nodes) 
    path = np.ones((m,nodes))

    for k in range(n):
        #initializing the first destination of all ants to the nearest node
        path[:,0]= cust_list.iloc[0]
    
        for i in range(m):
            # copy of path cost
            t_eta=np.array(eta)
        
            for j in range(nodes-1):
            
                #current node of the ant
                cur_loc = int(path[i,j]-1)       
                #setting cost of current node as zero
                t_eta[:,cur_loc] = 0     
            
                #calculating pheromone and cost for probablity evaluation
                p_feature = np.power(pheromone[cur_loc,:],beta)          
                c_feature = np.power(t_eta[cur_loc,:],alpha)  
            
                #creating 1D array of features
                p_feature = p_feature[:,np.newaxis]                     
                c_feature = c_feature[:,np.newaxis]                     
            
                #calculating the probability feature
                prob_feature = np.multiply(p_feature,c_feature)    
                probability = prob_feature/np.sum(prob_feature) 
            
                #calculating cumulative probability
                cum_prob = np.cumsum(probability)
            
                #choosing random number between [0,1]
                random.seed(0.5)
                r = np.random.random_sample()
            
                #finding next customer location having probability higher than random(r)
                cust_loc = np.nonzero(cum_prob>r)[0][0]+1      
            
                #adding customer location to the path 
                if cust_loc in cust_list.values:
                    path[i,j+1] = cust_loc                
            
            #finding and adding the last untraversed customer location in the path
            last = list(set([i for i in range(1,nodes+1)])-set(path[i,:-2]))[0]  
            if last in cust_list.values:
                path[i,-2] = last    
        
        #intializing optimal route   
        optimal_path = np.array(path)               
    
        #intializing total_distance_of_path with zero
        dist_cost = np.zeros((m,1))              
    
        #calcualting total tour distance
        #storing distance of tour for 'i'th ant at location 'i'
        for i in range(m):
            s = 0
            for j in range(nodes-1):
                s = s + distance[int(optimal_path[i,j])-1,int(optimal_path[i,j+1])-1]   
            dist_cost[i]=s           
        
        #finding location of minimum of dist_cost
        dist_min_loc = np.argmin(dist_cost)
        #finging min of dist_cost
        dist_min_cost = dist_cost[dist_min_loc]         
    
        #intializing current traversed as best route
        best_route = path[dist_min_loc,:]    
    
        #evaporating the pheromone levels
        pheromone = (1-decay)*pheromone  
    
        #updating the pheromone levels with delta tau. delta tau will be higher on min_dist
        for i in range(m):
            for j in range(nodes-1):
                dt = 1/dist_cost[i]
                pheromone[int(optimal_path[i,j])-1,int(optimal_path[i,j+1])-1] = pheromone[int(optimal_path[i,j])-1,int(optimal_path[i,j+1])-1] + dt   

    print('All paths generated :\n',optimal_path)
    print('Cost of best path',int(dist_min_cost[0]) + distance[int(best_route[-2])-1,0])
    print('Best Route:',best_route)
    return best_route
```

We then create a modified nearest neighbour function suitable for problem 2 which has been extended to also return the distance to the nearest customer.

```python
# Create a function to find the nearest customer that still needs fuel
# Takes 1D array of distances from a particular node indexed by node position &
# a copy of the same arrow sorted by distance
# Returns node id of nearest customer that needs fuel and the distance to that customer
def nearest_customer2(row_distance, sorted_distance):
    counter = 1
    customer_found = False
    
    # loop whilst customer hasn't been found and index is less than the number of nodes
    while counter < len(row_distance) and customer_found == False:
        index = 0
        is_nearest = False
        nearest_distance = 0
        
        # loop to find the nearest node
        while is_nearest == False and index < len(sorted_distance):
            if sorted_distance[counter] == row_distance[index]:
                nearest_distance = sorted_distance[counter]
                is_nearest = True
            else:
                index += 1

        # loop through cust_dict2 to check whether the node is a customer and needs fuel
        for i in cust_dict2:
            if index == i and cust_dict2[i]['fuel_needed'] > 0:
                customer_found = True
        if customer_found == False:
            counter += 1
    
    if counter == len(row_distance):
        return 0
    else:
        return index, nearest_distance;
```

```python
# Make a dictionary giving each low gas customer a value for its closest depot

low_cust_dict = {}

for cust in low_cust:
    depot_dist = []
    for depot in depot_id_list:
        depot_dist.append(dist_graph[cust, depot])
    min_index = depot_dist.index(min(depot_dist))
    low_cust_dict[cust] = depot_id_list[min_index]
    
low_cust_dict
```

We can then rearrange this dictionary so the depots are the keys and the values are a list of the low gas customers belonging to them.

```python
# Rearrange dictionary so keys are depot and values are list of low gas customers

depot_low_cust_dict = {}

depot_low_cust_dict[116] = []
depot_low_cust_dict[124] = []
depot_low_cust_dict[373] = []
depot_low_cust_dict[523] = []

for cust in low_cust_dict:
    depot_low_cust_dict[low_cust_dict[cust]].append(cust)

depot_low_cust_dict
```

Not all of the customers are able to be reached with the 250 mile distance contraint, so we narrow down the dictionary to only include customers which can be reached.

```python
# Narrow down which customers are reachable with the 250 mile distance constraint including a return to depot

depot_low_cust_dict_reachable = {}

depot_low_cust_dict_reachable[116] = []
depot_low_cust_dict_reachable[124] = []
depot_low_cust_dict_reachable[373] = []
depot_low_cust_dict_reachable[523] = []

for depot in depot_low_cust_dict:
    for cust in depot_low_cust_dict[depot]:
        if dist_graph[depot, cust] < 125:
            depot_low_cust_dict_reachable[depot].append(cust)

depot_low_cust_dict_reachable
```

From the un-optimised solution, we can see that only two of the low gas customers which can be reached have not been delivered to.

So we can manually deliver to them as the first stop for a lorry from each of their closest depots.

```python
# Only 2 of the low gas customers which are reachable are not delivered to in initial solution
# Customer 459 belonging to depot 116 & customer 497 belonging to depot 373
# So we manually deliver to them and then run greedy solution again from this state

# 116: [459]
# 373: [497]

lorries_dict2['116-0']['dist_travelled'] += dist_graph[116, 459]
lorries_dict2['116-0']['position'] = 459
lorries_dict2['116-0']['num_stops'] += 1
lorries_dict2['116-0']['level'] -= round(cust_dict2[459]['fuel_needed'], 2)
cust_dict2[459]['level'] += round(cust_dict2[459]['fuel_needed'], 2)
lorries_dict2['116-0']['fuel_delivered'] += round(cust_dict2[459]['fuel_needed'], 2)
solution_list2[23]['loc'].append([459, round(cust_dict2[459]['fuel_needed'] * -1, 2)])
cust_dict2[459]['fuel_needed'] = 0

lorries_dict2['373-0']['dist_travelled'] += dist_graph[373, 497]
lorries_dict2['373-0']['position'] = 497
lorries_dict2['373-0']['num_stops'] += 1
lorries_dict2['373-0']['level'] -= round(cust_dict2[497]['fuel_needed'], 2)
cust_dict2[497]['level'] += round(cust_dict2[497]['fuel_needed'], 2)
lorries_dict2['373-0']['fuel_delivered'] += round(cust_dict2[497]['fuel_needed'], 2)
solution_list2[13]['loc'].append([497, round(cust_dict2[497]['fuel_needed'] * -1, 2)])
cust_dict2[497]['fuel_needed'] = 0
```

Here we can see lorries '373-0' and '116-0' have had their first step of the journey manually initialised.

Now we can run the algorithm again to fill out the rest of the solution list.

```python
# Set index for solution file
sol_index2 = 0

# Set end statement
all_lorries_finished = False

# While end statement not True
while all_lorries_finished == False:
    # for each lorry
    for lorry in lorries_dict2:
        if sol_index2 > 24:
            sol_index2 = 0

        # if lorry finished journey, skip to next lorry
        if lorries_dict2[lorry]['finished_journey'] == True:
            sol_index2 += 1
            continue;
            
        # if lorry empty return to depot and fuel up, add to dist_travelled and num_stops
        if lorries_dict2[lorry]['level'] == 0:
            lorries_dict2[lorry]['dist_travelled'] += dist_graph[lorries_dict2[lorry]['position'], lorries_dict2[lorry]['depot']]
            lorries_dict2[lorry]['num_stops'] += 1
            lorries_dict2[lorry]['position'] = lorries_dict2[lorry]['depot']
            lorries_dict2[lorry]['level'] = lorries_dict2[lorry]['capacity']
            solution_list2[sol_index2]['loc'].append([lorries_dict2[lorry]['position'], lorries_dict2[lorry]['level']])

        # find closest unfilled customer
        # Make a 1D array of distances from a particular node
        row_distance2 = dist_graph[lorries_dict2[lorry]['position']]
        # Make a copy of this array sorted by distance
        sorted_distance2 = np.sort(row_distance2)
        # Find nearest unfilled customer and the distance to it
        unfilled_customer, customer_distance = nearest_customer2(row_distance = row_distance2, sorted_distance = sorted_distance2)
        if unfilled_customer == 0:
            break;

        # Check distance travelled by lorry won't go over 250 miles or number of stops isn't 5
        # If over 250 miles or at 5 stops then return to depot and set finished_journey to True
        if lorries_dict2[lorry]['dist_travelled'] + customer_distance + dist_graph[unfilled_customer, lorries_dict2[lorry]['depot']] > 250 or lorries_dict2[lorry]['num_stops'] == 5:
            lorries_dict2[lorry]['dist_travelled'] += dist_graph[lorries_dict2[lorry]['position'], lorries_dict2[lorry]['depot']]
            lorries_dict2[lorry]['position'] = lorries_dict2[lorry]['depot']
            solution_list2[sol_index2]['loc'].append([lorries_dict2[lorry]['position'], 0])
            lorries_dict2[lorry]['finished_journey'] = True
            sol_index2 += 1
            continue;
        else:
            # Set lorry position to unfilled customer and add to dist_travelled and num_stops
            lorries_dict2[lorry]['dist_travelled'] += customer_distance
            lorries_dict2[lorry]['position'] = unfilled_customer
            lorries_dict2[lorry]['num_stops'] += 1

        # If the lorry has less fuel than the customer fuel needed, deliver all and set lorry level to 0
        if lorries_dict2[lorry]['level'] < cust_dict2[unfilled_customer]['fuel_needed']:
            cust_dict2[unfilled_customer]['fuel_needed'] -= round(lorries_dict2[lorry]['level'], 2)
            cust_dict2[unfilled_customer]['level'] += round(lorries_dict2[lorry]['level'], 2)
            lorries_dict2[lorry]['fuel_delivered'] += round(lorries_dict2[lorry]['level'], 2)
            solution_list2[sol_index2]['loc'].append([unfilled_customer, round(lorries_dict2[lorry]['level'] * -1, 2)])
            lorries_dict2[lorry]['level'] = 0
        else: # If the lorry has equal or greater than fuel needed, deliver amount and set fuel needed to 0
            lorries_dict2[lorry]['level'] -= round(cust_dict2[unfilled_customer]['fuel_needed'], 2)
            cust_dict2[unfilled_customer]['level'] += round(cust_dict2[unfilled_customer]['fuel_needed'], 2)
            lorries_dict2[lorry]['fuel_delivered'] += round(cust_dict2[unfilled_customer]['fuel_needed'], 2)
            solution_list2[sol_index2]['loc'].append([unfilled_customer, round(cust_dict2[unfilled_customer]['fuel_needed'] * -1, 2)])
            cust_dict2[unfilled_customer]['fuel_needed'] = 0

        # Increment sol_index2
        sol_index2 += 1
    
    # Check end statement
    lorries_finished = True
    
    for lorry in lorries_dict2:
        if lorries_dict2[lorry]['finished_journey'] == False:
            lorries_finished = False
            break;
    
    if lorries_finished == True:
        all_lorries_finished = True
```

Again we can simply optimise the solution list by starting each lorry with only the amount of fuel it delivers.

### 5.1 Critical Evaluation

#### 5.1.1 Citical Evaluation for Greedy Search Algorithm used for Problem 1

The initial unoptimized solution for problem 1 delivered the total demand of 90.42 litres of gas to all customers but at a high cost of £155,002.50. With the simple optimization of initialising the lorries only with the amount of fuel they delivered on their journeys, we end up saving £106,265.63 making the total cost of this solution £48,736.87. This is a low cost, but this cost could be minimized further with several changes to the algorithm.

One possible way would be to have prioritised completing full journeys with each lorry at a time, rather than one step of the journey for each lorry at a time. By doing single steps of the journey, we risk lorries sharing the same node space and then making large jumps in distance to find the next closest node. In addition to this we could have prioritised the order in which lorries completed their journeys by capacity, sending the largest capacity lorries first and ending with the smallest. This could allow the large lorries to deliver to entire groups of customers in an area meaning that no other lorries would need to travel there. Finally, splitting the customers into 4 groups by assigning customers to each depot and only delivering to customers local to their depot could have potentially minimised costs further. Alternatively, this could be minimized by using a more intelligent algorithm such as a genetic algorithm or ant colony optimisation.

#### 5.1.2 Critical Evaluation for Genetic Algorithm used for Problem 1

Major complication faced while working on Genetic algorithm were:   
1. Performing crossover on the initial population leading to overlapping of the customer location.
2. Building fitness function incorporating the constraints (minimizing cost and considering fuel delivered).

#### 5.1.3 Critical Evaluation for Ant Colony Optimization Algorithm used for Problem 1

The ACO algorithm returned optimal routes to navigate, however the delivery plan could be optimised much better and below discussed are the 2 main shortcomings in our approach.

* Since the navigation on the ACO generated route cannot be completed by one lorry, the location at which one lorry reaches its capacity, rest of the journey is continued by the next lorry available in the depot. The subsequent lorries joining at the location last completed by previous lorry does not make for an ideal solution, as they could be missing a nearest location or the lorry could start a new shorter journey to deliver remaining customers.


* We created a group for each depot location to deliver to its nearest customers, the cluster creation was done using dijaktra’s multi source algorithm. The grouping did fair division amongst the depot locations based on distance. However, this was not a good enough distribution for our problem since each depot locations had different number of lorries, so the depot with highest number of lorries remained unused after having delivered to its respective customers while depots with smaller number of lorries made multiple trips to complete the delivery thereby increasing the costs. This could be optimised to make the division based on both distance and no. of lorries allocated to each depot for better delivery distribution.

#### 5.1.4 Critical Evaluation for Modified Greedy Search Algorithm used for Problem 2

The initial solution for problem 2, already optimised by only initialising each lorry with the amount of gas they deliver, produced a valid solution delivering 25.62 litres of gas at a cost of £21,587.93. After optimising the solution by delivering to all low gas customers first, in an attempt to reduce fines, we produced a valid solution with a total cost of £20,089.16 whilst delivering 27.54 litres of gas. This gives a total cost saving of £1,498.77 whilst delivering 1.92 more litres of gas.

This solution successfully delivers to all possible low gas customers with a cost of around £729.45 per litre delivered. As the total demand for gas is 90.42 litres, there is a lot of room to optimise further. Possible optimisations could be similar steps to what is highlighted for the greedy solution for problem 1, completing full journeys for each lorry instead of single steps, prioritising sending lorries by descending capacity and grouping customers by their nearest depot. Alternatively, producing a more intelligent algorithm such as a multi-objective genetic algorithm or multi-objective ant colony optimisation.
